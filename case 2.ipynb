```python
import pandas as pd
```

```python
# baca data csv menggunakan fungsi pada pandas
data = pd.read_csv('Mall_Customers.csv')

# tampilkan data
data
```

```python
# Case 2.2
'''
index pada data merupakan data dari 0 hingga 199. Anda diminta untuk mengubah index menjadi angka 
hasil penjumlahan 1000 dengan 'CustomerID'. Misal CustomerID = 9, index menjadi 1009. kemudian 
hapus kolom 'CustomerID'
'''

# membuat list data customer berdasarkan kolom customer ID
id_customer = data['CustomerID'].values + 1000

# ganti index menjadi list customer yang telah dibuat
data.index = id_customer

# tampilkan data
data
```

```python
# hapus kolom customer ID
data1 = data.drop('CustomerID', axis=1)

# tampilkan data
data1
```

```python
# case 2.3
'''
customer 1098 dan 1099 tidak lagi menjadi pelanggan mall tsb.
Anda diminta untuk menghapus data index 1098 dan 1099.
'''

# hapus baris dengan index 1098 dan 1099
data2 = data1.drop([1098, 1099], axis = 0)

# menampilkan data2
data2
```

```python
# case 2.4
'''
Ada data baru pada data customer.
Anda harus menambhakan data baru berikut.
'''

# membuat dataframe dari data
new_row = pd.DataFrame({
    'Genre': 'Male',
    'Age': 22,
    'Annual_Income_(k$)': 20,
    'Spending_Score': 83},
    index = [1201])

# tampilkan data baru
new_row
```

```python
# menambahkan data baru ke data yang telah dibuat sebelumnya
data3 = data2._append(new_row)

# menampilkan data3
data3
```

```python
data3.dtypes
```

```python
# case 2.5
'''
data income dan score kedepanya berpontensi berisi data float.
ubah data menjadi float
'''
# casting kolom Annual_Income_(k$)
data4 = data3.copy()

data4['Annual_Income_(k$)'] = data4['Annual_Income_(k$)'].astype(float)

# casting kolom Spending_Score
data4['Spending_Score'] = data4['Spending_Score'].astype(float)

# tampilkan type data tiap kolom
data4.dtypes

```

```python
# Exercise
## no 1
'''
ubah masing-masing header sesuai data berikut:
Genre = JenisKelamin
Age = Umur
Annual_Income_(k$) = Pendapatan
Spending_Score = Skor
'''

# buat dictionary nama header lama dan baru
new_header = {
    'Genre': 'JenisKelamin',
    'Age': 'Umur',
    'Annual_Income_(k$)': 'Pendapatan',
    'Spending_Score': 'Skor'
}

# ganti nama kolom menggunakan fungsi rename
data4.rename(columns= new_header, inplace=True)

# tampilkan data
data4
```

```python
# no 3
## ubah umur data index 1200 menjadi 31 dan update data index 1002 menjadi male, 22, 20, 83

# ubah umur data index 1200 dan 1002
# ubah umur index 1200
data5 = data4.copy()

data5.loc[[1200],['Umur']] = 31

# ubah data index 1002
data5.loc[[1002]] = ['Male', 22, 20, 83]
data5
```

```python
# no 4. Export data menjadi file excel
# export menggunakan fungsi to_excel()
data5.to_excel('Data_Pelanggan.xlsx')
```
